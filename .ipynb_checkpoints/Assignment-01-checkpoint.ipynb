# A Personal Example Illustrating the Difference Between Analysis and Analytics

In my academic and professional journey, I have often encountered situations where I had to work with data. At first, my instinct was always to look back at what happened and try to understand it. This backward-looking process is what I now know as analysis. Later, as I gained more experience, I began to see how data could also be used to look forward—to anticipate trends, predict outcomes, or optimize future decisions. That forward-looking process is analytics. Both are closely connected, but they serve very different purposes.

One of the clearest personal examples that illustrates this difference comes from my professional experience as an Assistant Manager at United Hospital. My role often required me to work with financial reports, budget data, and departmental expense summaries. These reports were full of numbers: revenues, expenses, supplier costs, maintenance fees, and operational charges. At first glance, they might look like just rows and columns in a spreadsheet, but I quickly learned that behind every figure was a story about the hospital’s operations and performance.

## The Situation: Unexpected Rise in Expenses

I remember one particular month when the hospital’s radiology department expenses were significantly higher than usual. As part of my role, I had to investigate this change and report my findings to senior management. The expense report showed that the costs were almost 20% higher than the average of the previous few months. The number itself was alarming, but numbers alone are never enough. To understand what was really going on, I had to dig deeper.

## How I Used Analysis

This is where analysis came into play. My first step was to compare the radiology department’s expenses with those from the previous three months. I broke the numbers down into categories: equipment maintenance, supplier payments, staff overtime, and consumables like films and contrast dyes used in imaging.

After carefully reviewing the invoices and usage records, I noticed two main reasons for the higher cost:

1. There had been a sudden increase in the number of MRI tests conducted that month.

2. One of the MRI machines had required urgent maintenance, which added an unexpected bill to the expenses.

By conducting this analysis, I was able to answer the key question: why did expenses increase this month? The analysis provided clarity, and I could confidently present the findings to management. The extra costs were not due to inefficiency or waste, but rather to higher patient demand and an unavoidable repair.

So, in simple words, analysis helped me look backward to understand what happened and why it happened.

## How Analytics Could Be Used

While analysis gave me answers for that month, I realized something important: the hospital was constantly reacting to unexpected expenses instead of anticipating them. That’s where analytics could have been valuable.

If we had used analytics, we could have taken several years of data on patient volumes, types of imaging tests, seasonal health trends, and machine maintenance records to predict future expenses. For example:

* By examining seasonal patterns, we might see that MRI demand tends to rise during winter due to an increase in accidents or    chronic pain cases.

* By analyzing machine usage data, we might predict when maintenance would likely be needed and set aside a budget in advance.

* By combining patient admission trends with test requests, we could forecast the likely number of imaging tests in upcoming months.

Using analytics, the hospital could move from being reactive to being proactive. Instead of being surprised by sudden expenses, we could prepare for them, allocate resources more efficiently, and negotiate better deals with suppliers.

So, in simple words, analytics would help us look forward to anticipate what might happen and prepare for it.

## An Analogy in My Own Words

The difference between analysis and analytics became very clear to me through this experience. I often explain it with a simple analogy:

Analysis is like looking in the rearview mirror of a car. You can clearly see the road you have already traveled, understand the bumps you hit, and explain why you had to slow down.

Analytics is like using a GPS. It takes information about where you are, combines it with maps and traffic data, and helps you decide the best route to reach your destination.

Both are essential. Without the rearview mirror, you can’t understand what has already affected your journey. Without the GPS, you can’t prepare for what lies ahead.

## A Broader Academic Example

A similar situation happened during my time as an R&D Officer at Independent University, where I worked with student feedback surveys. Each semester, students rated their courses and provided comments about their learning experience. One semester, we noticed that some courses consistently had lower satisfaction scores compared to others.

Through analysis, I reviewed the comments, compared the scores, and identified the main issues: heavy workloads and unclear instructions. This explained why students were less satisfied in those courses.

But if we had applied analytics, we could have gone further. By looking at several semesters of data—combining course structures, faculty teaching styles, and student performance—we could have predicted which courses might face similar challenges in the future. That would have allowed the university to adjust teaching methods, balance workloads, or provide faculty training before problems arose.

## A Personal Life Example

Even in my personal life, I have seen this distinction. For example, I once tracked my monthly expenses to understand where my money was going. By doing analysis, I found that most of my spending was on food and transportation. This helped me understand why I was not saving enough money.

However, by using analytics, I could have gone further by forecasting my future spending patterns based on my lifestyle, upcoming plans, and seasonal habits (like traveling more in summer). With that, I could predict when my expenses would likely rise and plan a savings strategy in advance.

Again, the pattern is the same: analysis explained the past, while analytics could guide the future.

## Why This Difference Matters

Through these experiences, I learned that while analysis and analytics are related, they serve different but complementary purposes. Analysis answers the question “What happened, and why?” It is backward-looking and descriptive. Analytics answers the question “What is likely to happen, and what should we do about it?” It is forward-looking and predictive or prescriptive.

In my hospital example, analysis allowed me to explain the sudden increase in expenses, while analytics could have helped prevent such surprises in the future. In my university example, analysis helped me identify problem areas in teaching, while analytics could have guided improvements for upcoming semesters. In my personal finance example, analysis explained why I wasn’t saving, while analytics could have helped me design a better financial plan.

## Reflection

Looking back, I see that most of us naturally perform analysis in our daily lives. We check our grades, bank statements, or work reports to understand what has already happened. But to move one step further—to prepare for the future, to optimize our decisions, and to reduce risks—we need analytics. It requires more data, better tools, and sometimes advanced techniques, but it transforms the way we use information.

For me, the lesson is simple: analysis is about learning from the past, while analytics is about shaping the future. Both are necessary, but analytics has the power to turn information into foresight.
